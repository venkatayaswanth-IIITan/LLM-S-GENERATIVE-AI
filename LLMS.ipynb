
# 🚀 PART 1: WHAT IS AN LLM?

## 1.1 Definition

An **LLM (Large Language Model)** is a deep learning model trained on **billions of words and sentences** to understand and generate human-like language.

It learns **patterns in text**, not fixed rules.

### Examples

* ChatGPT
* Gemini
* Claude
* Llama
* Qwen
* Mistral

---

## 1.2 Why is it called "Large"?

Because:

* It is trained on **huge datasets** (books, websites, articles, code, research papers).
* It contains **millions or billions of parameters**.
* It requires massive computational resources to train.

Example:

| Model            | Approx. Parameters |
| ---------------- | ------------------ |
| GPT-2 Small      | 117 Million        |
| Llama 3 8B       | 8 Billion          |
| GPT-class models | Billions+          |

---

## 1.3 Core Idea: Next Token Prediction

An LLM predicts **the next most likely token** based on previous tokens.

Example:

```
Input:
I love eating

Prediction:
I love eating biryani.
```

Generation process:

```
Prompt:
I love

↓

Predict:
eating

↓

Predict:
biryani

↓

Predict:
every

↓

Predict:
Sunday

↓

Final:
I love eating biryani every Sunday.
```

The model repeats this process until it decides to stop.

# 1.10 Common Misconceptions

❌ "LLMs think like humans."

➡️ They compute probabilities over sequences of tokens.

❌ "LLMs memorize every answer."

➡️ They learn patterns and representations, though they may memorize some training examples.

❌ "LLMs always tell the truth."

➡️ They can produce incorrect or fabricated information ("hallucinations").

# 3.1 What is a Token?

A **token** is the **smallest unit of text** that an LLM processes.

A token **is not always a word**. It can be:

* A complete word
* Part of a word (subword)
* A single character
* A punctuation mark
* A number
* A special symbol

### Example

```text
Sentence:
I love ChatGPT!

Possible Tokens:
["I", "love", "Chat", "G", "PT", "!"]
```

The exact split depends on the tokenizer used.

---

# 3.2 Why Don't LLMs Process Whole Words?

Suppose an LLM only stored complete words.

It would need separate entries for:

```text
run
running
runner
runs
ran
```

Instead, it can split them into reusable pieces:

```text
run + ning
run + ner
run + s
```

This:

* ✅ Reduces vocabulary size
* ✅ Handles unknown words better
* ✅ Improves efficiency

---

# 3.3 Real-Life Analogy

Think of **LEGO blocks**.

Instead of storing every possible building, you store reusable blocks and combine them.

Similarly:

```text
un + happy
play + ing
read + able
```

The model combines smaller pieces to understand new words.


Disadvantages:


Advantages:

* Easy to understand

Disadvantages:

* Unknown words become difficult.

---

## C. Subword Tokenization ⭐ (Used by Modern LLMs)

```text

```

Advantages:

* Handles unseen words
* Smaller vocabulary
* Better generalization

# 3.14 Context Window Uses Tokens

Suppose a model has:

```
Context Window = 8 tokens
```

Input:

```

```

The tokenizer might produce 10+ tokens.

If the context limit is exceeded:

* Older tokens may be dropped, or

This is why context windows are measured in **tokens**, not characters or words.

# 🚀 PART 4: What Are Embeddings? (Complete Interview Notes)

> **Goal:** Understand what embeddings are, why LLMs need them, how they are created, and how they help models understand the meaning of text.

---

# 4.1 What Are Embeddings?

An **embedding** is a **numerical vector (list of decimal numbers)** that represents the meaning of a word, sentence, or document.

👉 In simple words:

> **Embeddings convert human language into numbers that AI models can understand while preserving semantic meaning.**

### Example

```text
Word: "Cat"

Embedding:
[0.23, -0.91, 0.45, 0.78, ..., 0.12]
```

The model **does not understand the word "Cat" directly**. It understands its **embedding vector**.

---

# 4.2 Why Do We Need Embeddings?

Neural networks **cannot process text directly**.

Example:

```text
"I love AI"
```

The computer cannot calculate with letters.

So we convert it into:

```text
"I love AI"
      │
      ▼
 Tokenization
      │
      ▼
["I", "love", "AI"]
      │
      ▼
 Token IDs
      │
      ▼
[40, 1842, 9552]
      │
      ▼
 Embeddings
      │
      ▼
[
 [0.12, 0.45, ...],
 [0.91, -0.21, ...],
 [0.67, 0.10, ...]
]
```

Now the Transformer can process the input mathematically.

---

# 4.3 Real-Life Analogy

Imagine a world map.

```
Hyderabad  ---- Bangalore
      \
       \
        Chennai
```

Cities that are geographically close are near each other on the map.

Similarly, in embedding space:

```
King ------ Queen

Man ------- Woman

Cat ------- Dog

Apple ----- Mango
```

Words with similar meanings are located close together.

---

# 4.4 Difference Between Token IDs and Embeddings

| Token ID                | Embedding                     |
| ----------------------- | ----------------------------- |
| Integer                 | Floating-point vector         |
| Example: `9552`         | Example: `[0.34, -0.72, ...]` |
| Just an index           | Contains semantic meaning     |
| Used to look up vectors | Used by the Transformer       |

# 7.1 What is a Transformer?

A **Transformer** is a deep learning architecture introduced in the 2017 paper **"Attention Is All You Need"**.

It processes **all input tokens in parallel** using **self-attention**, making it much faster and more effective than RNNs and LSTMs.

### In Simple Words

> **Transformer = A neural network architecture that understands relationships between all words in a sentence simultaneously.**

---

# 7.2 Why Were Transformers Introduced?

## ❌ Problem with RNN

```text
I → Love → Learning → AI
```

* Processes one word at a time
* Slow
* Difficult to remember long contexts

---

## ❌ Problem with LSTM

LSTM improved memory but still:

* Sequential processing
* Hard to parallelize
* Expensive for very long sequences

---

## ✅ Transformer Solution

```text
             I
           ↙ ↓ ↘
      Love  AI  Learning
         ↖  ↓  ↗
```

Every token can interact with every other token **at the same time**.

---

```text
                     ▼

        │   Transformer Block 1   │
        └─────────────────────────┘
                     │
        ┌─────────────────────────┐
        │   Transformer Block 2   │

                     │

                     │
        ┌─────────────────────────┐
```

Every token becomes a high-dimensional representation.

# 7.7 Step 3 – Positional Encoding

## Why?

Transformers process tokens **in parallel**.

Without extra information:

```text

Man bites dog
```

could appear identical.

So positional information is added.

### Example

```text
Token      Position

Dog            1
bites          2
man            3
```

The embedding is updated:

```text
Final Input =
Token Embedding
+
Position Embedding
```

This allows the model to understand word order.

This is the **heart of the Transformer**.

Sentence:

```text

```

Question:

Who is **He**?

Self-attention learns:

```text
He ─────────► Virat Kohli
```

instead of incorrectly connecting to "century".

# 8.1 What is Attention?

The **Attention Mechanism** allows a model to **focus on the most relevant words or tokens** when understanding or generating text.

### Simple Definition

> **Attention tells the model which words are important and which words should receive more focus.**

---

# 8.2 Real-Life Analogy 🎯

Imagine you are reading this sentence:

> **"Rohit gave the ball to Virat because he was injured."**

When you read **"he"**, your brain automatically tries to determine **who "he" refers to**.

Humans naturally pay attention to the relevant words.

LLMs do exactly the same thing using the **Attention Mechanism**.


Question:

Possible references:

* Trophy ✅
* Suitcase ❌

With attention:

```text
                it
                 │
                 ▼
```

The model learns that **"it" refers to "trophy"**.


Problems:

Every word can look at every other word.

```text

       ↘  ↓  ↙
          mat
```

This enables richer understanding of relationships.

# 8.10 Attention Score

The Query is compared with every Key.

Example:

```text
Sentence:

```

When processing:

```text
cat
```

Scores:

```text
cat      → 0.95

mouse    → 0.60

the      → 0.10

chased   → 0.75
```

Higher score = more attention.

# 8.11 Softmax

Raw scores:

```text
[3.2, 1.8, 0.4]
```

Softmax converts them into probabilities:

```text
[0.73, 0.21, 0.06]
```

Properties:

* All values are between 0 and 1.
* They sum to 1.

---

# 8.12 Weighted Sum

Each Value vector is multiplied by its attention weight.

```text
Attention Weight × Value

↓

Weighted Values

↓

Sum

↓

Final Representation
```

This becomes the updated representation of the token.

# 9.17 Conceptual Training Loop (Pseudo-code)

```python

    # Convert text to tokens
    inputs = tokenizer(batch)

    # Model prediction

    optimizer.step()

    # Clear old gradients
    optimizer.zero_grad()
```

# 12.9 Step 7 – Logits

After the Transformer finishes processing, it produces **logits**.

Example:

```text
Prompt:

The capital of France is

Logits:

Paris   → 18.4
London  →  2.1
Rome    →  0.9
Berlin  →  0.7
```

Logits are **raw scores**, not probabilities.

---

# 12.10 Step 8 – Softmax

Softmax converts logits into probabilities.

Example:

| Token  | Probability |
| ------ | ----------- |
| Paris  | 0.97        |
| London | 0.02        |
| Rome   | 0.005       |
| Berlin | 0.005       |

These probabilities sum to **1.0**.


# 12.14 Autoregressive Generation

ChatGPT generates text like this:

```text
Prompt

↓

Predict Token 1

↓

Append

↓

Predict Token 2

↓

Append

↓

Predict Token 3

↓

Append

↓

Continue

↓

Stop
```

This is called **autoregressive decoding**.

# 12.23 Interview Questions

### Q1. How does ChatGPT answer a question?

**Answer:**
It tokenizes the input, converts tokens into embeddings, processes them through Transformer layers with self-attention, computes logits, converts them into probabilities using Softmax, selects the next token using a decoding strategy, appends it to the sequence, and repeats until completion.

---

### Q2. Does ChatGPT generate an entire paragraph in one step?

**Answer:**
No. It generates **one token at a time** using autoregressive decoding.

---

### Q3. What happens after tokenization?

**Answer:**
Tokens are mapped to token IDs, transformed into embeddings, and processed by Transformer layers.

---

### Q4. What is the role of Softmax?

**Answer:**
Softmax converts raw logits into a probability distribution over possible next tokens.

---

### Q5. Does inference change the model's weights?

**Answer:**
No. During inference, the model performs forward computation only and does not update its parameters.
